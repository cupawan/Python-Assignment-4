```python
class triangle():
    def __init__(self,side1,side2,side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        
    def cal_s(self):
        s= (self.side1 + self.side2 + self.side3)/2
        return s
        
```

```python
class area_triangle(triangle):
    def __init__(self,side1,side2,side3):
        super().__init__(side1,side2,side3)
        
    def area(self):
        s = (self.side1+self.side2+self.side3)/2
        area = (s*(s-self.side1)*(s-self.side2)*(s-self.side3))**0.5
        print("Area of the triangle is {} square units.".format(round(area,3)))
```

```python
#Demonstration of above program:
t1 = area_triangle(5,6,9)
t1.area()
```

```python
def filter_long_words(l,n):
    long_words = []
    for i in l:
        if len(i)>n:
            long_words.append(i)
    return long_words            
```

```python
#Demonstration of the above function:
list1 = ["sun","earth","tree","glass","two","childern","iNeuron","ML","python","van","row"]
filter_long_words(list1,4)
```

```python
evens = ["a","e","i","o","u"]
def check_even(x):
    if x in evens:
        return True
    else:
        return False
```

```python
# Demonstration of the above function:
print(check_even('e'))
print(check_even('z'))
```
